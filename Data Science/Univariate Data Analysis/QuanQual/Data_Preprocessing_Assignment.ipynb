```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np

import warnings
warnings.filterwarnings("ignore")
from Univariate import Univariate

dataset = pd.read_csv("kidney_disease.csv")
# dataset.dtypes
# quantitative - Numeric columns
# Qualitative - String columns
quan,qual = Univariate.quanQual(dataset)
print(f"Quantitive columns: {quan}")
print(f"Qualitative columns: {qual}")

# # dataset['salary'].fillna(0, inplace=True)
# # dataset.isnull().sum()
quan_df = Univariate.CreateImputerForReplaceNan("Quan", dataset, quan)
print(f"Quantitative columns check: \n{quan_df.isnull().sum()}")
qual_df = Univariate.CreateImputerForReplaceNan("Qual", dataset, qual)
print(f"Qualtative columns check: \n{qual_df.isnull().sum()}")
pre_processed_df = pd.concat([quan_df, qual_df], axis=1)

print(f"Final PreProcessed data check: \n{pre_processed_df.isnull().sum()}")
pre_processed_df.to_csv("kidney_disease_pre_processed_data.csv",index=False)

```
