```python
from Univariate import Univariate
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

dataset = pd.read_csv("Placement.csv")

quan,qual = Univariate.quanQual(dataset)
dataset.describe()

descriptive = pd.DataFrame(index=["Q1:25%","Q2:50%","Q3:75%","Q4:100%","IQR","1.5Rule","Lesser","Greater","Min","Max"], columns=quan)
descriptive_res = Univariate.measureOfLocation(descriptive,dataset, quan)
print(descriptive_res)

lesser_outlier_list, greater_outlier_list = Univariate.findingOutlier(descriptive, quan)
print(f"Lesser outlier list: {lesser_outlier_list}")
print(f"Greater outlier list: {greater_outlier_list}")
```

```python
ds = Univariate.replaceTheOutlier(descriptive, dataset, lesser_outlier_list, greater_outlier_list)
# print(f"DataSet Table :{ds}")
descriptive = pd.DataFrame(index=["Q1:25%","Q2:50%","Q3:75%","Q4:100%","IQR","1.5Rule","Lesser","Greater","Min","Max"], columns=quan)
descriptive_res = Univariate.measureOfLocation(descriptive,ds, quan)
lesser_outlier_list, greater_outlier_list = Univariate.findingOutlier(descriptive, quan)
print(f"Lesser outlier list: {lesser_outlier_list}")
print(f"Greater outlier list: {greater_outlier_list}")
```

```python
#Finding outlier
lesser = list()
greater = list()
for column in quan:
    if descriptive[column]["Min"] < descriptive[column]["Lesser"]:
        lesser.append(column)
    if descriptive[column]["Max"] > descriptive[column]["Greater"]:
        greater.append(column)
     
print(f"Lesser outlier column list: {lesser}")
print(f"Greater outlier column list: {greater}")
```

```python
# Replace the outliner
for columnName in lesser:
    dataset[columnName][dataset[columnName]<descriptive[columnName]["Lesser"]] = descriptive[columnName]["Lesser"]

for columnName in greater:
    dataset[columnName][dataset[columnName]<descriptive[columnName]["Greater"]] = descriptive[columnName]["Greater"]

```

```python
descriptive = pd.DataFrame(index=["Q1:25%","Q2:50%","Q3:75%","Q4:100%","IQR","1.5Rule","Lesser","Greater","Min","Max"], columns=quan)
for column in quan:
    descriptive[column]["Q1:25%"]= dataset.describe()[column]["25%"]  #np.percentile(dataset[column],25)
    descriptive[column]["Q2:50%"]= dataset.describe()[column]["50%"]  #np.percentile(dataset[column],50)
    descriptive[column]["Q3:75%"]= dataset.describe()[column]["75%"]  #np.percentile(dataset[column],75)
    descriptive[column]["Q4:100%"]= dataset.describe()[column]["max"]
    descriptive[column]["IQR"]= descriptive[column]["Q3:75%"] - descriptive[column]["Q1:25%"]
    descriptive[column]["1.5Rule"]= 1.5*descriptive[column]["IQR"]
    descriptive[column]["Lesser"]= descriptive[column]["Q1:25%"]-descriptive[column]["1.5Rule"]
    descriptive[column]["Greater"]= descriptive[column]["Q3:75%"]+descriptive[column]["1.5Rule"]
    descriptive[column]["Min"]= dataset[column].min()
    descriptive[column]["Max"]= dataset[column].max()
descriptive
#Finding outlier
lesser = list()
greater = list()
for column in quan:
    if descriptive[column]["Min"] < descriptive[column]["Lesser"]:
        lesser.append(column)
    if descriptive[column]["Max"] > descriptive[column]["Greater"]:
        greater.append(column)
     
print(f"Lesser outlier column list: {lesser}")
print(f"Greater outlier column list: {greater}")
```
